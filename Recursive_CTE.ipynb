#Recursive CTE란?
recursive CTE 또는 재귀쿼리는 자기 자신을 참조하여 결과를 생성한다.

전체 결과 집합을 얻을 때까지 데이터의 하위 집합을 반복적으로 실행하는데, 
이를 통해 데이터를 계층적으로 표현 할 수 있다.

정확하진 않지만 쉽게말하자면 파이썬의 for 문을 SQL에서 구현 하는 방법이다.

#Recursive CTE의 문법
with RESCURSIVE TABLENAME AS (  
  쿼리1(기준 쿼리)  
  UNION ALL (중복 값 허용) - (**중복값 허용 X 시 UNION 사용**)  
  쿼리2(반복 쿼리)  
)  
SELECT *  
FROM TABLENAME  

---
## 과정
RECURSIVE CTE가 실행되면 기준 쿼리인 **쿼리1** 이 시행되고,  
 UNION ALL 을 통해 반복쿼리인 **쿼리 2**가 계속 합쳐지는 구조이다.  
 마지막으로 이전 호출에서 반환 할 결과가 없다면 RECURSIVE CTE가 종료된다.

```python
#문제 : 24시간을 나타내는 열을 만드시오
%%bigquery --project ringed-marker-347607
WITH RECURSIVE TEMP AS (
    SELECT 0 AS HOUR
    UNION ALL
    SELECT HOUR+1 FROM TEMP
    WHERE HOUR<23
)
SELECT *
FROM TEMP
ORDER BY HOUR;
```
